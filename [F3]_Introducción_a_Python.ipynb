## Variables y aritmética básica
Declaramos variables usando el `=` (por ejemplo `a=3`). A izquierda siempre debe estar el nombre de la variable, pero a derecha puede haber una expresión tan compleja como quiera (por ejemplo, `a=3+2*7**9`). Además de los operadores aritméticos básicos `+`, `-`, `*`, `/` tenemos la potencia `a**b` (equivalente a $a^b$) y los operadores de comparación:

Op. | Descripción
--- | ---
`==` | `True` si son iguales
`!=` | `True` si son distintos
`>` | `True` si la izquierda es mayor a la derecha
`<` | `True` si la izquierda es menor a la derecha
`>=` | `True` si la izquierda es mayor o igual a la derecha
`<=` | `True` si la izquierda es menor o igual a la derecha

Aquí `True` es un tipo de dato conocido cómo _bool_, asociado a un valor de verdad (en este caso, _Verdadero_). Naturalmente, tiene su opuesto `False` (_Falso_) que también puede definirse cómo `not True` (el `not` es el operador _negación_).

```
# Creo las variables a y b y me divierto operando
a = 2
b = 5
# Los caracteres entre comillas dobles o simples se toman como literales (strings)
# La función print escribe en pantalla las variables que recibe
# (puede recibir múltiples variables, siempre separadas por comas)
print("Sumo a+b:", a+b)
print("Resto a-b:", a-b)
print("Multiplico a*b:", a*b)
print("Divido b/a:", b/a)
print("División entera b//a:", b//a)
print("Resto de división b%a:", b%a)
print("Comparo a>b:", a>b)
print("Comparo a==b:", a==b)
print("Comparo a<=b:", a<=b)
```

## Funciones, `if` y `while`

Todas estas estructuras tienen un _cuerpo_ de líneas que ejecutan bajo condiciones específicas. El cuerpo debe indentarse: cada línea debe comenzar con un `Tab` (↹).

Las _funciones_ son subrutinas que realizan cómputos y devuelven un resultado usando `return`. Se definen usando `def` seguido de su nombre identificador y una lista de argumentos entre paréntesis. El cuerpo se ejecuta cada vez que son invocadas.

Los `if` y `while` se definen utilizando un _bool_, cuyo valor determina la acción.

* `if` ejecuta el cuerpo una única vez si el _bool_ es `True`. Puede además agregarse un `else` que ejecuta su cuerpo solo si el `if` no ejecutó el suyo (cuando el _bool_ es `False`).

* `while` repite la ejecución de su cuerpo mientras el bool sea `True`. Esto significa que un `while(True)` nunca termina. ¡Ojo!


```
#  Defino una función que eleve al cuadrado
print("Funcion al_cuadrado:")
def al_cuadrado(x): # 'al_cuadrado' es el nombre y x la variable
  x_cuadrado = x**2
  return x_cuadrado

x1 = 3
x1_cuad = al_cuadrado(x1) # Se reemplaza por lo que esta junto al "return"
print("x1_cuad =", x1_cuad)
print("al_cuadrado(2)=", al_cuadrado(2))

# Quiero que me escriba todos los numeros hasta el 8 (sin incluir)
print("\nEscribo todos los numeros del 1 al 7")
num = 1
while num < 8:  # Mientras el número que estoy mirando sea menor a 8
  print(num)    # lo printeo por pantalla
  num = num + 1 # y paso al siguiente (si no hacemos esto, ¡nunca termina!)

# Ahora quiero que me diga si son pares o no
# El if solo ejecuta su cuerpo si su condicion es True, sino ejecuta el cuerpo de else
# (no es necesario que haya else, en cuyo caso no hace nada si la condicion es False)
print("\nEscribo todos los numeros del 1 al 7, indicando su paridad")
num = 1
while num < 8:
  if num % 2 == 0: # Operador resto, ver bloque anterior
    print(num, "par")
  else: # Si no es par, debe ser impar
    print(num, "impar")
  num = num + 1
```

## Listas y `for`

Las _listas_ son colecciones ordenadas de variables (de cualquier tipo), que se escriben separados por comas y entre 2 corchetes.
Se utilizan para almacenar información y son útiles para guardar grandes cantidades de datos en una única variable.
Puedo acceder al _i-esimo_ elemento de una lista `A` con `A[i]`. Recuerden que los índices comienzan en `0` por lo que el primer elemento sería `A[0]`.
La longitud de la lista se obtiene con `len(A)`.

Las listas tienen _constructores_, funciones diseñadas para crear listas rápidamente cómo por ejemplo `range(n,m,s)` que devuelve la lista \\
`[n, n+s, n+2*s, ..]` sin incluir `m`.
También pueden utilizarse ciclos para construir listas a mano utilizando la función `append`, que agrega un elemento al final de la lista.

Similarmente al `while`, el `for` recorre una lista y ejecuta el cuerpo para cada elemento de ella.

**Aclaración:** Las listas no son ni pueden usarse cómo vectores, no son apropiados para hacer cuentas; para eso veremos el `array` de NumPy. Tomenlas cómo un contenedor de información y no (mucho) más.

```
mi_lista = [3.14, -8, "F3", True, [1,23]]
print("Mi lista es", mi_lista, "y tiene longitud", len(mi_lista))
print("\nLa recorro con un for y printeo cada elemento")
for elemento in mi_lista:
  print(elemento)

print("\nPrinteo un range(1, 8, 1)")
lista = list(range(1, 11, 1))
print(lista)
for i in range(1, 8, 1): # Range genera todos los numeros entre 1 y 8 (sin incluir) saltando de a 1
  print(i)
# Sirve para solo los elementos pares, por ejemplo
print("\nAhora solo los pares de mi lista inicial")
for i in range(0, len(mi_lista), 2): # Recorro los indices entre 0 y 5 (sin incluir) saltando de a 2
  print(mi_lista[i]) # Accedo al i-esimo elemento de mi_lista (¡que será siempre par!)

# Uso un ciclo para construir una lista con todos los cubos entre 1**3 y 8**3 (sin incluir)
mis_cubos = []
i = 1
while i < 8:
  mis_cubos.append(i**3) # Agrego nuevo elemento al final
  i = i + 1
print("\nLista con los primeros 8 cubos perfectos, usando while")
print(mis_cubos)
# Esto también se puede hacer con un for, ¡prueben!
mis_cubos = [i**3 for i in range(1, 8, 1)] # Lista por comprension
# Equivalente
mis_cubos = []
for i in range(1, 8, 1):
  mis_cubos.append(i**3)
```

Podemos definir un array a mano, pero numpy posee múltiples funcionalidades para crear arrays útiles inmediatamente (similares al `range`). En el fondo son listas pero más limitadas, especializadas en operaciones algebráicas y analíticas.

**Cualquier operación básica sobre arrays se realiza elemento a elemento**. Esto incluye las funciones matemáticas que trae NumPy como `np.cos`, `np.log`, `np.arctanh`, etc.
Por lo tanto, no necesito un `while` para calcular el coseno de cada uno de los 100 elementos de un array `x`, me basta con utilizar `np.cos(x)`.

```
# Si no le digo a Python que lo quiero usar, no puedo acceder a sus funciones,
# pero lo rebauticé np porque es más corto
import numpy as np

# Creo arrays, uso el prefijo np. para que sepa que son funciones de numpy
A = np.array([-1, 2, 3.14]) # Transformo si lista en un array
B = np.zeros(3) # Crea un vector de 3 ceros
print("A =", A)
print("B =", B)
print("Producto elemento a elemento A*B =", A*B)
print("Suma elemento a elemento A+B =", A+B)
# Me aburro! Pongamosle algún valor no nulo a B...
B[0] = 3 # Seteo el primer elemento (¡se cuenta desde 0!)
print("Potencia elemento a elemento A**B =", A**B)
print("Le aplico al_cuadrado a B:", al_cuadrado(B))
```

## Matrices y como jugar con muchos indices

Cómo dijimos, un array puede tener un rango (cantidad de índices) arbitrario. Con rango 2, tenemos una matriz `M`, lo cual cambia un poco cómo operan los índices.
Un elemento ahora de identifica con 2 indices _i_, _j_ accediendo cómo `M[i, j]`.

NumPy nos permite crear a _slices_ (cortes) de una matriz usando `M[i,:]`, que devuelve un vector `A` (rango 1) tal que `A[j] == M[i, j]`.
Pero podemos devolver _slices_ acotando el tamaño. Por ejemplo, `M[i, n:m:s]` devuelve un array los elementos `M[i, n]`, `M[i, n+s]`, `M[i, n+2*s]`, .. de forma análoga a un `range`.

```
import numpy as np

M = np.zeros((3, 4)) # Crea una matriz nula de 3x4
print("Matriz nula\n", M)
print("de dimension ", M.shape)
# Le pongo un valor en las esquinas
M[0, 0] = 8
M[0, 3] = 8
M[2, 0] = 8
M[2, 3] = 8
print("\nPoniendo 8 en las esquinas queda \n", M)
# Le pongo un único valor a toda la 2da fila
M[1, :] = -3
print("\nCambiando toda la 2da fila queda \n", M)
# Lleno los 0's de la primera fila con -10
M[0, 1:-1] = -10 # Un índice negativo se lee cómo N-1 donde N es la longitud
print("\nCambiando parte de la 1ra fila queda \n", M)
# Le pido a M la 2da fila y su 3era columna
print("\nMediante slices accedo a...")
print("Segunda fila:", M[1, :])
print("Tercera columna:", M[:, 2])
```

##**Generación automática de arreglos y máscaras**

Cómo ya vimos, NumPy dispone de múltiples funciones para generar arreglos de tamaño y dimensión arbitrarios de forma automática, siguiendo alguna regla.



*   `np.linspace(x0, x1, N, endpoint=True)`: Devuelve un arreglo equiespaciado de $N$ valores entre `x0` y `x1`. Si `endpoint` es `True` (valor por defecto), incluye `x1`. Sino, no lo incluye.
*   `np.arange(x0, x1, dx)`: Devuelve un arreglo con los valores entre `x0` y `x1` (sin incluir) equiespaciados en `dx`; `[x0, x0+dx, x0+2*dx, ...]`. A diferencia del `linspace`, **en `arange` definimos el paso, no la cantidad de puntos.**
*   `np.ones(shape)`: Devuelve un arreglo con las dimensiones definidas por `shape` lleno de 1s.
*   `np.zeros(shape)`: Devuelve un arreglo con las dimensiones definidas por `shape` lleno de 0s.
*   `np.random.random(shape)`: Devuelve un arreglo con las dimensiones definidas por `shape` lleno de números pseudoaleatorios entre 0 y 1.

Cómo vimos, podemos generar arreglos con reglas más complicadas aplicando [funciones matemáticas](https://numpy.org/doc/stable/reference/routines.math.html) a los arreglos generados por estas reglas simples.

Además, podemos usar _máscaras_ para acceder a elementos específicos de un arreglo o para hacer operaciones más binarias (ver ejemplos).
Una máscara es un arreglo de booleanos `True`/`False` que puede obtenerse aplicando operadores lógicos a un arreglo.

```
import numpy as np
np.set_printoptions(precision=3) # Magia negra para que todo
np.random.seed(0)                # salga y se vea lindo

# Creo algunos arreglos de juguete
A = np.random.random((10))
B = np.random.random((10))
C = np.ones((10))/2
print("Construyo una máscara que me dice en que elementos es B mayor que A")
A_menor_B = A < B
print(" A =  ", A)
print(" B =  ", B)
print("mask1=", A_menor_B, "\n")
print("Ahora construyo otra máscara que me diga donde A es mayor que 1/2 (C)")
A_menor_05 = A >= C
print(" A =  ", A)
print("mask2=", A_menor_05, "\n")

print("Ahora la intersección (and) y disjuncion (or) de ambas mascaras")
print("       mask1 =  ", A_menor_B)
print("       mask2 =  ", A_menor_05)
print("mask1 or  mask2=", A_menor_B | A_menor_05)
print("mask1 and mask2=", A_menor_B & A_menor_05, "\n")

print("Finalmente, uso la máscara para acceder a los elementos que cumplen la condicion")
print("Elementos de A menores a su contraparte en B:", A[A_menor_B])
print("Elementos de A menores a su 0.5:             ", A[A_menor_05])
print("Elementos de A que cumplen alguna de las 2:  ", A[A_menor_B | A_menor_05])
print("Producto de A por su máscara(True=1, False=0)", A*A_menor_B)
```

## **Matplotlib**

Matplotlib es un paquete cargado con funciones que permiten graficar... _lo que sea_. Vamos a empezar por algo simple, la función más básica de matplotlib es `plot(x, y, args*)` que grafica el array `y` en función del array `x`. Los `args*` son muchos argumentos opcionales que me permiten hacer el gráfico más lindo (color, tipo y grosor del trazo, por ejemplo).

Para graficar una función f(x), primero debemos calcular el array `y=f(x)` y luego hacer `plot(x, y, args*)`. Para graficar en un intervalo [a, b] con `N` puntos equiespaciados, definimos la variable `x=np.linspace(a, b, N)`.

Pero matplotlib tiene mucho más; podemos graficar campos vectoriales, lineas de campo, histogramas, mapas de calor, superficies/curvas en 3D. En la página de matplotlib pueden encontrar detalladas las distintas funciones: https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

Igual no se preocupen, en alguna clase futura lo veremos!


```
# Incluyo matplotlib para poder graficar
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_rara(x):
  y = x*np.cos(x)*np.exp(-x/25)
  return y

Xs = np.linspace(-np.pi, np.pi, 1000) # Quiero un espaciado bien fino
Ys = mi_funcion_rara(Xs) # Calculo el array para poder graficar la funcion
plt.figure() # Abro una figura vacía
plt.plot(Xs, Ys) # Grafico!
plt.show() # Le pido que me muestre
```

## **Arreglos multidimensionales**

¿Que pasa si ahora queremos graficar funciones de $R^2\to R$ (potenciales) o $R^2\to R^2$ (campos)?
Cómo antes, tenemos que discretizar nuestro dominio, en este caso un cuadrado, usando `np.linspace` para $x$ e $y$ (no necesariamente iguales):

`Xs=np.linspace(x_min, x_max, N)`

`Ys=np.linspace(y_min, y_max, M)`

Pero para dividir este cuadrado en celditas, necesitamos $N\times M$ pares $(x,y)$ y yo tengo $N$ de uno y $M$ de otro. Para hacer esto necesitamos un *producto cartesiano* entre `Xs` e `Ys`, que me devuelva matrices `X_m` e `Y_m` de $N\times M$. Estas matrices van a tener un montón de valores repetidos, ¡pero así me aseguro que todos los pares $(x,y)$ estén!

Esto lo hace la función `np.meshgrid()` que puede recibir una cantidad arbitraria de vectores 1D y, para cada uno, devuelve su tensor correspondiente obtenido del producto cartesiano.

```
import numpy as np
# Arranquemos con una discretización simple
Xs = np.linspace(0, 10, 11)
Ys = np.linspace(0, 5, 6)
print("Las versiones vectoriales son:")
print("Xs =", Xs)
print("Ys =", Ys)
X_m, Y_m = np.meshgrid(Xs, Ys) # Devuelve 2 argumentos, uno por cada vector que le pasé
print("\nLas versiones matriciales son:")
print("X_m =\n", X_m)
print("Y_m =\n", Y_m)

print("\nUsando slices, vemos que X_m tiene cómo filas Xs")
print("X_m[i,:] =", X_m[0,:], "= Xs")
print("y que Y_m tiene cómo columnas Ys")
print("Y_m[:,j] =", Y_m[:, 0], "= Ys")
```

## Campos escalares y vectoriales

Teniendo ya una forma de definir los pares $(x,y)$, nos interesa ahora definir una función $f:R^2\to R$. Si tengo $N$ valores de $x$ y $M$ valores de $y$, almacenados en 2 matrices de $N\times M$, es razonable pensar que los valores de $f(x,y)$ también estarán almacenados en una matriz de $N\times M$. Para un campo vectorial $\bar{F}(x,y) = (F_x(x,y), F_y(x,y))$ simplemente tendré 2 de estas matrices, una por componente.

Tomemos por ejemplo el campo eléctrico de una carga puntual en el plano $x$-$y$:
\begin{equation*}
\bar{E} = \frac{kq}{r^3}\bar{r} = kq\frac{(x\hat{x}+y\hat{y})}{(x^2+y^2)^{3/2}} \\
E_x(x,y) = kq\frac{x}{(x^2+y^2)^{3/2}} \\
E_y(x,y) = kq\frac{y}{(x^2+y^2)^{3/2}} \\
\end{equation*}

En el siguiente bloque definimos las funciones $E_x$ e $E_y$ en Python.
En principio, operamos cómo si las variables de entrada `x` e `y` fueran números, pero vamos a aprovechar que *NumPy opera sobre arrays elemento a elemento*. De esta forma, **los x e y que les pasemos a la función pueden ser matrices y su resultado serán también matrices**.

```
import numpy as np
import matplotlib.pyplot as plt

def E_x(x, y): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2)
  return x/r**3

def E_y(x, y): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2)
  return y/r**3

Xs = np.linspace(-2, 2, 200) # Así construidos, me aseguro que el (0,0) no
Ys = np.linspace(-2, 2, 200) # esté en mi dominio (el campo diverge ahí!)
X_m, Y_m = np.meshgrid(Xs, Ys)
Ex = E_x(X_m, Y_m)
Ey = E_y(X_m, Y_m)

plt.streamplot(Xs, Ys, Ex, Ey, density=2)
plt.plot([0], [0], "ko", markersize=10) # Pongo un punto en el centro representando la carga
plt.xlabel("x")
plt.ylabel("y")
plt.title("Campo de una carga puntual")
```

## Vectoriales: Streamplots, mapas de calor y contours

Arriba vimos el ejemplo de un `streamplot`, que grafica las _líneas de campo_ en 2D. Esta es una forma muy buena de visualizar la dirección del campo, pero no da mucha información sobre el módulo. Esto es muy bueno para comprender las deformaciones del campo al superponer configuraciones.

Los mapas de calor, por otro lado, permiten visualizar una función de $R^2\to R$ mediante un gradiente de colores: es ideal para graficar _potenciales electrostáticos_. Para esto usamos la función `plt.imshow` que recibe una matriz y asigna un color a cada punto dependiendo del valor de la matriz en ese punto. Esto solo funciona para matrices equiespaciadas y requiere que le pongamos los límites a mano.

Por otro lado, la función `plt.contours` nos permite graficar las superficies equipotenciales; donde $f(x,y)=cte$. Esto complementa muy bien el mapa de calor y nos permite ver cómo las líneas de campo son perpendiculares.

Cómo ejemplo, vamos a aplicar todo esto a un sistema de 2 cargas $q$ y $-q$ a distancia $d$

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, SymLogNorm

# Campo y potencial de carga una puntual, con desplazarlos d/2 y cambiarles el
# signo puedo obtener el campo de cada carga
def E_x(x, y): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2)
  return x/r**3

def E_y(x, y): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2)
  return y/r**3

def pot(x, y): # Tomo k*q=1 para simplificar
  r = np.sqrt(x**2+y**2)
  return 1/r

Xs = np.linspace(-2, 2, 2000)
Ys = np.linspace(-2, 2, 2000)
X_m, Y_m = np.meshgrid(Xs, Ys)

d = 1

# Campos de la carga en (-d/2, 0), sumo d/2 en x
E1_x = E_x(X_m+d/2, Y_m)
E1_y = E_y(X_m+d/2, Y_m)
V1 = pot(X_m+d/2, Y_m)
# Campos de la carga en (d/2, 0), sumo -d/2 en x
E2_x = -E_x(X_m-d/2, Y_m) # Pongo un - porque es negativa
E2_y = -E_y(X_m-d/2, Y_m)
V2 = -pot(X_m-d/2, Y_m)
# El campo total es la suma de los campos
Ex = E1_x + E2_x
Ey = E1_y + E2_y
V = V1 + V2

plt.figure(figsize=(8, 8))
# Streamplot (pongo linea color verde)
plt.streamplot(Xs, Ys, Ex, Ey, color="g", density=1)


# Mapa de calor, uso extent para definir los límites x-y del mapa
# Uso el gradiente de color rainbow porque está lindo y una escala
# logarítmica para el gradiente de calor
plt.imshow(np.abs(V), extent=[-2, 2, -2, 2], cmap = plt.cm.rainbow, norm=LogNorm())
#plt.colorbar(label=r"$|V(x,y)|$", size=22) # Pongo la barra que indica que color es que valor
plt.colorbar().set_label(label=r"$|V(x,y)|$",size=22)

# Curvas de nivel, en levels le digo que valores de f(x,y) quiero
CS = plt.contour(V, extent=[-2, 2, -2, 2],
                linewidths=2, colors = "black", levels=[-3, -1, -0.5, -0.2, 0, 0.2, 0.5, 1, 3])
# Le agrego los valores f(x,y) a cada curva
plt.clabel(CS, colors = "black", inline=True, fmt="%.1f", fontsize=14)

# Extras estéticos
plt.plot([-d/2], [0], "bo", markersize=12) # Pongo puntos representando las cargas
plt.plot([d/2], [0], "ro", markersize=12)
plt.xlabel("x", fontsize=20)
plt.ylabel("y", fontsize=20)
plt.title("Campo de un dipolo no puntual", fontsize=22)
plt.show()
```

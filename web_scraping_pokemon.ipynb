#Guia de Web Scraping

Escribimos el siguiente código para manejar el pandas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Escribimos la dirección url donde vamos a extraer la información

```python
url="https://pokemondb.net/pokedex/all"
r=requests.get(url)
soup=BeautifulSoup(r.content, 'html.parser')
```

En el siguiente código le especificamos que encuentre la tabla con el id pokedex, que encuentre el tbody y que al final encuentre y tome todo lo tr que se encuentran en el tbody.

```python
rows=soup.find('table',attrs={"id":"pokedex"}).find('tbody').find_all('tr')
```

Aquí le podemos indicar que queremos encontrar un elemento determinado.

```python
rows[0].find_all('td')[1].get_text()
```

Mostramos la información de los pokemon

```python
names=[]
for row in rows:
    names.append(row.find_all('td')[1].get_text())
```

```python
names
```

Si queremos saber la cantidad de filas utilizamos len

```python
len(names)
```

Ahora queremos saber de que tipo es cada pokemon

```python
rows[0].find_all('td')[1].get_text()
rows[0].find_all('td')[2].get_text()
```

Ahora si queremos ver la todos los tipos imprimo types

```python
names=[]
types=[]
for row in rows:
    names.append(row.find_all('td')[1].get_text())
    types.append(row.find_all('td')[2].get_text())
```

```python
types
```

Ahora sabiendo esto de como descomponer cada lista creamos nuestro dataframe

```python
names=[]
types=[]
total=[]
hp=[]
attacks=[]
defense=[]
sp_attacks=[]
sp_def=[]
sppeds=[]
for row in rows:
    names.append(row.find_all('td')[1].get_text())
    types.append(row.find_all('td')[2].get_text())
    total.append(row.find_all('td')[3].get_text())
    hp.append(row.find_all('td')[4].get_text())
    attacks.append(row.find_all('td')[5].get_text())
    defense.append(row.find_all('td')[6].get_text())
    sp_attacks.append(row.find_all('td')[7].get_text())
    sp_def.append(row.find_all('td')[8].get_text())
    sppeds.append(row.find_all('td')[9].get_text())
```

```python
df=pd.DataFrame({"nombres":names,"types":types,"total":total,"hp":hp,"attack":attacks,"defense":defense,"sp_attack":sp_attacks,"sp_def":sp_def,"spped":sppeds})
```

```python
df
```

Ahora escribimos el código para crear mi archivo csv.

```python
df.to_csv('pokemones.csv')
```
